```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = 'https://www.imdb.com/chart/top/'
```

```python
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'}
```

```python
response = requests.get(url, headers=headers)
```

```python

if response.status_code == 200 :
    soup = BeautifulSoup(response.text, 'html.parser')
else :
    print('Failed to retrieve the webpage. Status code:', response.status_code)
```

```python
movies = soup.find_all('div', class_='ipc-title ipc-title--base ipc-title--title ipc-title-link-no-icon ipc-title--on-textPrimary sc-1e00898e-9 jQixeG cli-title')
```

```python
year = soup.find_all('div', class_='sc-1e00898e-7 hcJWUf cli-title-metadata')
```

```python
rating = soup.find_all('span', class_='ipc-rating-star ipc-rating-star--base ipc-rating-star--imdb ratingGroup--imdb-rating')
```

```python
Name = []
Rating = []
No_of_Rating = []
Year = []
Time = []
```

```python
for i in range(0,len(movies)) :
    print('---------------------------------------------')
    print()
    print('#. ',i+1)
    n_m = movies[i].text.split('. ')[1]
    r_t = rating[i].text.strip().replace(')','').split('\xa0(')
    if len(year[i].text) > 3 :
        y_r = year[i].text[:4]
        if 'h' in year[i].text[4:] :
            h = year[i].text[4:].split('h')[0]
            if 'm' in year[i].text[4:].split('h')[1] :
                m = year[i].text[4:].split('h')[1].split('m')[0]
            else :
                m = ''
        else :
            h = ''
            m = ''
    else :
        y = ''
        h = ''
        m = ''
    h_m = h + ' h' + m +' m'
    
    print('Name : ',n_m)
    print('Rating : ',r_t[0])
    print('No of Rating : ',r_t[1])
    print('Year : ',y_r)
    print('Time : ',h_m)
    
    print()
    
    Name.append(n_m)
    Rating.append(r_t[0])
    No_of_Rating.append(r_t[1])
    Year.append(y_r)
    Time.append(h_m)
```

```python
df = pd.DataFrame()
```

```python
df['Name'] = Name
df['Rating'] = Rating
df['No of Rating'] = No_of_Rating
df['Year'] = Year
df['Time'] = Time
```
