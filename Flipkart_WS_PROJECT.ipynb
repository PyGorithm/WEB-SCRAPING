```python
pd_name = []
pd_price = []
pd_offer = []
pd_rv_link = []
pd_rating = []
pd_no_rating = []
pd_no_reviews = []
pd_5_str = []
pd_4_str = []
pd_3_str = []
pd_2_str = []
pd_1_str = []
```

```python
for url in url_arr :
    response = requests.get(url)
    if response.status_code == 200 :
        soup = BeautifulSoup(response.text, 'html.parser')
        
        print('------------------------------------------------------------')
        print()
        print()
        
        product_name_tag = soup.find('span', class_='B_NuCI')
        if product_name_tag:
            product_name = product_name_tag.text.strip()
            pd_name.append(product_name)
            print('Product Name:', product_name)
            print("")
            
        price_tag = soup.find('div', class_='_30jeq3 _16Jk6d')
        if price_tag:
            price = price_tag.text.strip()
            pd_price.append(price)
            print('Price:', price)
            print("")
            
        offer_elements = soup.find_all('li', class_='_16eBzU')
        
        offers = []
        for offer in offer_elements:
            offer_text = offer.text.strip()
            offer_s = offer_text + ' , '
            
        for index, offer in enumerate(offers, start=1):
            print(f'Offer {index}: {offer}')
            
        pd_offer.append(offer_s)
            
        print()
            
        comment_links = soup.find_all('a', href=True)
        
        if comment_links :
            comment_links_filtered = [link['href'] for link in comment_links if 'comment' in link['href'] or 'review' in link['href']]
            urls = 'https://www.flipkart.com'+comment_links_filtered[-1]
            
            print(urls)
            pd_rv_link.append(urls)
            print()
            
            responses = requests.get(urls)
            
            if responses.status_code == 200 :
                soups = BeautifulSoup(responses.text, 'html.parser')
                
                rating_elements = soup.find_all('div', class_='_2d4LTz')
    
                if rating_elements:
                    rating_text = rating_elements[0].text.strip()
                    pd_rating.append(rating_text)
                    print("Overall Rating : ",rating_text)
                    print("")
                    
                no_ratings_reviews = soup.find_all('div', class_='row _2afbiS')
                
                if no_ratings_reviews :
                    ratings_text = no_ratings_reviews[0].text.strip()
                    print('Ratings :', ratings_text.split(" ")[0])
                    pd_no_rating.append(ratings_text.split(" ")[0])
                    
                    print("")

                    reviews_text = no_ratings_reviews[1].text.strip()
                    print('Reviews :', reviews_text.split(" ")[0].replace(',',''))
                    pd_no_reviews.append(reviews_text.split(" ")[0].replace(',',''))
                    print("")
                    
                no_5_str = soup.find_all('div', class_='_1uJVNT')
                if no_5_str :
                    no_5_star = no_5_str[0].text.strip()
                    print('5 Star :', no_5_star)
                    pd_5_str.append(no_5_star)


                    no_4_star = no_5_str[1].text.strip()
                    print('4 Star :', no_4_star)
                    pd_4_str.append(no_4_star)


                    no_3_star = no_5_str[2].text.strip()
                    print('3 Star :', no_3_star)
                    pd_3_str.append(no_3_star)


                    no_2_star = no_5_str[3].text.strip()
                    print('2 Star :', no_2_star)
                    pd_2_str.append(no_2_star)


                    no_1_star = no_5_str[4].text.strip()
                    print('1 Star :', no_1_star)
                    pd_1_str.append(no_1_star)
                    
                
            
            else :
                print('Failed to retrieve the webpage. Status code:', response.status_code)
                
        print()
        print()
            
        
        
    else :
        print('Failed to retrieve the webpage. Status code:', response.status_code)
```

```python
df = pd.DataFrame()
df['Name'] = pd_name 
df['Price'] = pd_price 
df['Offers'] = pd_offer 
df['Review Link'] = pd_rv_link 
df['Rating'] = pd_rating 
df['No of Rating'] = pd_no_rating 
df['No Of Reviews'] = pd_no_reviews 
df['5 Star'] = pd_5_str
df['4 Star'] = pd_4_str 
df['3 Star'] = pd_3_str
df['2 Star'] = pd_2_str 
df['1 Star'] = pd_1_str 
```
